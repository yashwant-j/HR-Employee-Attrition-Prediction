```python
#Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
#Get the number of rows and number of columns in the data
df.shape
```

```python
#Get the column data types
df.dtypes
```

```python
#Count the empty (NaN, NAN, na) values in each column
df.isna().sum()
```

```python
#Another check for any null / missing values
df.isnull().values.any()
```

```python
#View some basic statistical details like percentile, mean, standard deviation etc.
df.describe()
```

```python
#Get a count of the number of employee attrition, the number of employees that stayed (no=0) and the number that left (yes=1)
df['Attrition'].value_counts()
```

```python
#Show the number of employees that left and stayed by age
import matplotlib.pyplot as plt
fig_dims = (12, 4)
fig, ax = plt.subplots(figsize=fig_dims)

#ax = axis
sns.countplot(x='Age', hue='Attrition', data = df, palette="colorblind", ax = ax,  edgecolor=sns.color_palette("dark", n_colors = 1));
```

```python
#Print all of the object data types and their unique values
for column in df.columns:
    
        print(str(column) + ' : ' + str(df[column].unique()))
        print(df[column].value_counts())
        print("_________________________________________________________________")
```

```python
# droping irrelevant data columns

#Removing ID column
df=df.drop('Id', axis = 1)
#Remove the column EmployeeNumber
df = df.drop('EmployeeNumber', axis = 1) # A number assignment  
#Remove the column Behaviour
df = df.drop('Behaviour', axis = 1) #Contains only the value 1
```

```python
#Transform non-numeric columns into numerical columns
from sklearn.preprocessing import LabelEncoder

for column in df.columns:
        if df[column].dtype == np.number:
            continue
        df[column] = LabelEncoder().fit_transform(df[column])
```

```python
#Get the correlation of the columns
df.corr()
```

```python
#Visualize the correlation
plt.figure(figsize=(20,20))  #14in by 14in
sns.heatmap(df.corr(), annot=True, fmt='.0%')
```

```python
# Shifting Age column to end of the data
Age = df['Age']
df = df.drop('Age', axis = 1)
df['Age'] = Age
df
```

```python
#scatter matrix
from pandas.plotting import scatter_matrix
p=scatter_matrix(df, alpha=0.2, figsize=(12, 12), diagonal='kde')
```

```python
#Split the data into independent 'X' and dependent 'Y' variables
X_train = df.iloc[:, 1:df.shape[1]].values 
Y_train = df.iloc[:, 0].values
```

```python
#Use Random Forest Classification algorithm
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
forest.fit(X_train, Y_train)
```

```python
#Get the accuracy on the training data
forest.score(X_train, Y_train)
```

```python
# preparing testing data
```

```python
X_test = X_train
Y_test = Y_train
```

```python
#Show the confusion matrix and accuracy for  the model on the test data
#Classification accuracy is the ratio of correct predictions to total predictions made.
from sklearn.metrics import confusion_matrix

pred= forest.predict(X_test)
cm = confusion_matrix(Y_test, pred)
  
TN = cm[0][0]
TP = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]
  
print(cm)
print('Model Testing Accuracy = "{}!"'.format(  (TP + TN) / (TP + TN + FN + FP)))
print()# Print a new line
```

```python
#Confusion Matrix

from sklearn.metrics import confusion_matrix, classification_report, precision_score
pred=forest.predict(X_test)
cm_df = pd.DataFrame(confusion_matrix(Y_test, pred).T, index=forest.classes_, columns=forest.classes_)
cm_df.index.name = 'Predicted'
cm_df.columns.name = 'True'
print(cm_df)
print()
print('Model score = ',forest.score(X_test,Y_test))
```

```python
from sklearn.metrics import roc_curve, auc
import pylab as pl

(fpr, tpr, thresholds) = roc_curve(Y_test,pred)
area = auc(fpr,tpr)
pl.clf() #Clear the current figure
pl.plot(fpr,tpr,label="ROC Curve with area = %1.2f"%area)

pl.plot([0, 1], [0, 1], 'k') #This plots the random (equal probability line)
pl.xlim([0.0, 1.0])
pl.ylim([0.0, 1.0])
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.legend(loc="lower right")
pl.show()
```

```python
# Return the feature importances (the higher, the more important the feature).
importances = pd.DataFrame({'feature':df.iloc[:, 1:df.shape[1]].columns,'importance':np.round(forest.feature_importances_,3)}) #Note: The target column is at position 0
importances = importances.sort_values('importance',ascending=False).set_index('feature')

#Visualize the importance
importances.plot.bar()
```
